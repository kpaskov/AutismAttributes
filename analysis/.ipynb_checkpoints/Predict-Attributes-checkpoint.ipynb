```python
import csv
from collections import defaultdict
from itertools import combinations
import random
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import statsmodels.api as sm
import sklearn.metrics as metrics
from sklearn.svm import LinearSVC
from sklearn.ensemble import ExtraTreesClassifier
```

# Load data

```python
class Child():
    def __init__(self, data):
        self.id = data[1]
        self.family = data[0]
        self.relatedness = data[2]
        self.sex = data[5]
        self.age = None if data[6]=='' or int(data[6])<0 or int(data[6])>100 else int(data[6])
        self.diagnoses = set(data[7].split(','))
        self.asd = data[8]
        self.premature = data[9]
        self.gestation = 40 if data[10]=='' else int(data[10])
        self.verbal = int(data[11])
        self.sleep = data[12]
        self.hearing = data[13]
        self.vision = data[14]
        self.noise = data[15]
        self.motor = data[16]
        self.metabolic = data[17]
        self.gi = data[18]
        self.infections = data[19]
        self.seizures = data[20]
        
    def __str__(self):
        return 'Child(id=%s, fam=%s)' % (self.id, self.family)

fam_to_children = defaultdict(list)
children = []
with open('../data/kfPedsFileExport_09-12-2022.csv', 'r') as f:
    reader = csv.reader(f)
    header = next(reader)
    for line in reader:
        child = Child(line)
        fam_to_children[child.family].append(child)
        children.append(child)
```

```python
child_to_index = dict([(x.id, i) for i, x in enumerate(children)])

# -1 is unknown; 0 is NT behavior; 1 is AUT behavior
scq = -np.ones((len(children), 40))

aut_behavior = ['0', '0'] + ['1']*6 + ['0'] + ['1']*9 + ['0']*22
missing_from_ped = list()

with open('../data/kidsFirstAdminExport_09-20-2022_SCQ.csv', 'r') as f:
    reader = csv.reader(f)
    
    # header
    next(reader)
    header = next(reader)
    scq_questions = header[20:60]
    next(reader)
    next(reader)
    next(reader)
    
    for line in reader:
        child = line[0]
        if child in child_to_index:
            scq[child_to_index[child], :40] = [-1 if x=='' else 1 if x==aut_behavior[i] else 0 for i, x in enumerate(line[20:60])]
        else:
            missing_from_ped.append(child)
            
print('missing from ped', len(missing_from_ped))
has_data = np.any(scq!=-1, axis=1)
print('missing from scq', np.sum(~has_data))
print('data complete', np.sum(has_data))
```

# Preprocessing

```python
# is the child verbal? (pulled from SCQ)
is_verbal = (scq[:, 0]==0)
print('removing %d nonverbal children' % np.sum(has_data & ~is_verbal))
print('removing %d children missing scq' % np.sum(~has_data))


# remove children missing scq and children who are not verbal (they don't have answers to some SCQ questions)
children = [x for i, x in enumerate(children) if has_data[i] and is_verbal[i]]
child_to_index = dict([(x.id, i) for i, x in enumerate(children)])

scq = scq[has_data & is_verbal, :]
assert np.all(scq>=0)

print(len(children), scq.shape)
```

```python
# check that scq==1 is more common in autistic children than in their neurotypical siblings
# we've changed scq questions so that 1 represents the autistic behavior and 0 represents the neurotypical behavior
is_asd = np.array([child.asd=='yes' for child in children])
assert np.all(np.sum(scq[is_asd]==1, axis=0)[1:]/np.sum(is_asd) > np.sum(scq[~is_asd]==1, axis=0)[1:]/np.sum(~is_asd))
```

```python
# asd: yes/no
# premature: yes/no/missing=''
# sleep: yes/no/missing=''
# hearing: yes_hearingaids/yes_other/no/missing=''
# vision: yes/yes_other/no
# noise: yes_severe/yes_moderate/yes_mild/yes_other/yes_prefers/no/missing=''
# motor: yes/yes_some/no
# metabolic: yes/no/missing=''
# gi: yes/no/''
# infections: yes_constantly/yes_often/yes_slightly/no/missing=''
# seizures: yes/no/missing=''
# gestation: 23-40
# verbal: 1-10

# one-hot encode attributes
one_hot_attributes = np.zeros((len(children), 20))
cat_attrs = ['asd', 'noise', 'motor', 'sleep', 'gi', 'premature', 'hearing', 'vision', 'metabolic', \
             'infections', 'seizures']

one_hot_labels = []
for attr in cat_attrs:
    options = [y for y in set([getattr(x, attr) for x in children]) if y!='' and y!='no']
    one_hot_attributes[:, len(one_hot_labels):(len(one_hot_labels)+len(options))] = np.array([[int(getattr(x, attr)==opt) for x in children] for opt in options]).T
    one_hot_labels.extend(['%s:%s' % (attr, opt) for opt in options])
    
```

```python
print(one_hot_labels)
```

```python
# create an offset vector
offset = np.ones((len(children), 1))

# split into train and test (10%) by family
is_train = np.ones((len(children),), dtype=bool)
for fam, cs in fam_to_children.items():
    if random.random() < 0.1:
        for child in cs:
            if child.id in child_to_index:
                is_train[child_to_index[child.id]] = False
            
print('train', np.sum(is_train), 'test', np.sum(~is_train))
```

# First, let's look at how using different features impacts performance

```python
# Here, we're using a logistic regression model, regularized with L1 regression (meaning we encourage the model to
# use as few features as possible). The regularization means we need to tune a parameter alpha, which we do through
# cross validation.

# this is our L1-regularized logistic regression model
def train_logreg_l1_model(y, data, alpha):
    model = sm.Logit(y, data)
    result = model.fit_regularized(method='l1_cvxopt_cp', alpha=alpha, disp=False)
    
    if not result.mle_retvals["converged"]:
        result = None
        
    return result

# here, we pick the alpha which produces the best AUROC (area under the ROC curve) in cross validation.
def tune_params(y, data, train_model, alphas=[0.005, 0.01, 0.05, 0.1, 0.5, 1, 5, 10, 50]):
    # 5-fold cross-validation
    folds = np.random.randint(0, 5, size=(len(y),))
    rocs = np.zeros((len(alphas), 5))
    nonzero_params = np.zeros((len(alphas), 5))
    for i, alpha in enumerate(alphas):
        for fold in range(5):
            result = train_model(y[folds!=fold], data[folds!=fold, :], alpha)
            if hasattr(result, 'params'):
                params = result.params
            else:
                params = result.coef_
            if result is not None:
                rocs[i, fold] = metrics.roc_auc_score(y[folds==fold], result.predict(data[folds==fold, :]))
                nonzero_params[i, fold] = np.sum(~np.isclose(params, 0))
            else:
                rocs[i, fold] = np.nan
                nonzero_params[i, fold] = np.nan
        print(alpha, nonzero_params[i, :])
        
    best_alpha = alphas[np.nanargmax(np.nanmedian(rocs, axis=1))]
    print(best_alpha, np.nanmedian(rocs, axis=1))
    
    return best_alpha
```

```python
# for every attribute, we train classifiers using only training data
alpha_v = np.zeros((len(cat_attrs)))
alpha_attr = np.zeros((len(cat_attrs)))
alpha_vattr = np.zeros((len(cat_attrs)))

X_vs = []
X_attrs = []
X_vattrs = []

ys = []

for i, attr in enumerate(cat_attrs):
    is_other_attr = np.array([not x.startswith(attr) for x in one_hot_labels])
    y = np.array([1 if 'yes' in getattr(child, attr) else 0 for child in children])
    ys.append(y)

    # model including only scq
    X = np.hstack((offset, scq[:, 1:]))
    alpha_v[i] = tune_params(y[is_train], X[is_train, :], train_logreg_l1_model)
    X_vs.append(X)

    # model including only other attributes
    X = np.hstack((offset, one_hot_attributes[:, is_other_attr]))
    alpha_attr[i] = tune_params(y[is_train], X[is_train, :], train_logreg_l1_model)
    X_attrs.append(X)
               
    # model including only scq + other attributes
    X = np.hstack((offset, scq[:, 1:], one_hot_attributes[:, is_other_attr]))
    alpha_vattr[i] = tune_params(y[is_train], X[is_train, :], train_logreg_l1_model)
    X_vattrs.append(X)
    
```

```python
# now plot the data
plt.figure(figsize=(20, 20))

for i, attr in enumerate(cat_attrs):
    result_attr = train_model(ys[i][is_train], X_attrs[i][is_train, :], alpha_attr[i])
    result_v = train_model(ys[i][is_train], X_vs[i][is_train, :], alpha_v[i])
    result_vattr = train_model(ys[i][is_train], X_vattrs[i][is_train, :], alpha_vattr[i])
    
    plt.subplot(6, 4, 2*i + 1)
    
    fpr, tpr, _ = metrics.roc_curve(ys[i][~is_train], result_attr.predict(X_attrs[i][~is_train, :]))
    plt.plot(fpr, tpr, label='attrs (%d features)' % np.sum(~np.isclose(result_attr.params, 0)))
        
    fpr, tpr, _ = metrics.roc_curve(ys[i][~is_train], result_v.predict(X_vs[i][~is_train, :]))
    plt.plot(fpr, tpr, label='scq (%d features)' % np.sum(~np.isclose(result_v.params, 0)))
    
    fpr, tpr, _ = metrics.roc_curve(ys[i][~is_train], result_vattr.predict(X_vattrs[i][~is_train, :]))
    plt.plot(fpr, tpr, label='scq+attrs (%d features)' % np.sum(~np.isclose(result_vattr.params, 0)))
    
    plt.title('%s ROC' % attr)
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.plot([0, 1], [0, 1], color='black', linestyle='--')
    plt.legend()
    plt.xlim([0, 1.1])
    plt.ylim([0, 1.1])

    plt.subplot(6, 4, 2*i + 2)

    precision, recall, _ = metrics.precision_recall_curve(ys[i][~is_train], result_attr.predict(X_attrs[i][~is_train, :]))
    plt.plot(recall, precision, label='attrs (%d features)' % np.sum(~np.isclose(result_attr.params, 0)))
        
    precision, recall, _ = metrics.precision_recall_curve(ys[i][~is_train], result_v.predict(X_vs[i][~is_train, :]))
    plt.plot(recall, precision, label='scq (%d features)' % np.sum(~np.isclose(result_v.params, 0)))
    
    precision, recall, _ = metrics.precision_recall_curve(ys[i][~is_train], result_vattr.predict(X_vattrs[i][~is_train, :]))
    plt.plot(recall, precision, label='scq+attrs (%d features)' % np.sum(~np.isclose(result_vattr.params, 0)))
    
    plt.axhline(np.sum(ys[i]==1)/len(ys[i]), color='black', linestyle='--')

    plt.title('%s Precision-Recall' % attr)
    plt.xlabel('recall')
    plt.ylabel('precision')
    plt.legend()
    plt.xlim([0, 1.1])
    plt.ylim([0, 1.1])

plt.tight_layout()
plt.show()
```

```python
# I interpret these plot to mean that 
# asd, noise, motor are highly predictable
# sleep, gi, infections are somewhat predictable (there's a little signal there)
# prematurity, vision, metabolic are not predictable

# hearing and seizures are somewhat predictable from other attributes, but not from SCQ
```

```python
# we can take a look at which features are used
for i, attr in enumerate(cat_attrs):
    result_scq = train_model(ys[i][is_train], X_vs[i][is_train, :], alpha_v[i])
    
    print('\n', attr)
    
    ordered_indices = np.flip(np.argsort(np.abs(result_scq.params)))
    for i in ordered_indices:
        if i!=0 and result_scq.params[i] != 0:
            print(result_scq.params[i], scq_questions[i])
    
```

# We can also compare different classifiers

```python
# Here, we're using a logistic regression model, regularized with L1 regression (meaning we encourage the model to
# use as few features as possible). The regularization means we need to tune a parameter alpha, which we do through
# cross validation.

# this is our model
def train_svm_l1_model(y, data, alpha):
    model = LinearSVC(penalty='l1', loss='squared_hinge', dual=False, C=alpha, fit_intercept=False)
    result = model.fit(data, y)
        
    return result
    

# for every attribute, we train different classifiers using only training data
alpha_logreg = np.zeros((len(cat_attrs)))
alpha_svm = np.zeros((len(cat_attrs)))

for i, attr in enumerate(cat_attrs):
    is_other_attr = np.array([not x.startswith(attr) for x in one_hot_labels])
    y = ys[i]
    X = X_vs[i]

    # logistic regression
    alpha_logreg[i] = tune_params(y[is_train], X[is_train, :], train_logreg_l1_model)

    # SVM
    alpha_svm[i] = tune_params(y[is_train], X[is_train, :], train_svm_l1_model)
```

```python
# now plot the data
plt.figure(figsize=(10, 30))

for i, attr in enumerate(cat_attrs):
    result_logreg = train_logreg_l1_model(ys[i][is_train], X_vs[i][is_train, :], alpha_logreg[i])
    result_svm = train_svm_l1_model(ys[i][is_train], X_vs[i][is_train, :], alpha_svm[i])
    result_tree = ExtraTreesClassifier(n_estimators=50).fit(X_vs[i][is_train, :], ys[i][is_train])
    
    plt.subplot(len(cat_attrs), 2, 2*i + 1)
    
    fpr, tpr, _ = metrics.roc_curve(ys[i][~is_train], result_logreg.predict(X_vs[i][~is_train, :]))
    plt.plot(fpr, tpr, label='logreg (%d features)' % np.sum(~np.isclose(result_logreg.params, 0)))
        
    fpr, tpr, _ = metrics.roc_curve(ys[i][~is_train], result_svm.decision_function(X_vs[i][~is_train, :]))
    plt.plot(fpr, tpr, label='svm (%d features)' % np.sum(~np.isclose(result_svm.coef_, 0)))
    
    fpr, tpr, _ = metrics.roc_curve(ys[i][~is_train], result_tree.predict_proba(X_vs[i][~is_train, :])[:, 1])
    plt.plot(fpr, tpr, label='tree (%d features)' % result_tree.n_features_in_)
        
    plt.title('%s ROC' % attr)
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.plot([0, 1], [0, 1], color='black', linestyle='--')
    plt.legend()
    plt.xlim([0, 1.1])
    plt.ylim([0, 1.1])

    plt.subplot(len(cat_attrs), 2, 2*i + 2)

    precision, recall, _ = metrics.precision_recall_curve(ys[i][~is_train], result_logreg.predict(X_vs[i][~is_train, :]))
    plt.plot(recall, precision, label='logreg (%d features)' % np.sum(~np.isclose(result_logreg.params, 0)))
        
    precision, recall, _ = metrics.precision_recall_curve(ys[i][~is_train], result_svm.decision_function(X_vs[i][~is_train, :]))
    plt.plot(recall, precision, label='svm (%d features)' % np.sum(~np.isclose(result_svm.coef_, 0)))

    precision, recall, _ = metrics.precision_recall_curve(ys[i][~is_train], result_tree.predict_proba(X_vs[i][~is_train, :])[:, 1])
    plt.plot(recall, precision, label='tree (%d features)' % result_tree.n_features_in_)

    plt.axhline(np.sum(ys[i]==1)/len(ys[i]), color='black', linestyle='--')

    plt.title('%s Precision-Recall' % attr)
    plt.xlabel('recall')
    plt.ylabel('precision')
    plt.legend()
    plt.xlim([0, 1.1])
    plt.ylim([0, 1.1])

plt.tight_layout()
plt.show()
```
